```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# Lets make a function for plotting data
def plot_sample(x, y, index):
    plt.figure(figsize = (15, 2))
    plt.imshow(x[index])
# printing label of each class
    plt.xlabel(classes[y[index]]) # y_index show the array of classes
```

```python
# Normalizing or scaling the data by dividing 255
x_train = x_train/ 255
x_train

x_test = x_test/ 255
x_test
```

First Build the ANN 

```python
# MOdel building
ANN_model = keras.Sequential([
    keras.layers.Flatten(input_shape = (32, 32, 3)), #32 * 32 * 3 = 3072 so no of hiddel layers less than input
    keras.layers.Dense(3000, activation = 'relu' ),
    keras.layers.Dense(1000, activation = 'relu' ),

    keras.layers.Dense(10, activation = 'sigmoid' ) 
    
# 100 is no of neuron you specify in hidden layers must less than input
# 10 is output vectors, inputs shape is 784 
])

ANN_model.compile(optimizer = 'SGD',  loss = 'sparse_categorical_crossentropy', metrics = ['accuracy']) # when you have one hot encoding values use categorical_crossentropy otherwise sparse
# sparse_categorical_crossentropy means our data is in digits not in hot encoding labels

ANN_model.fit(x_train, y_train, epochs = 10)
```

```python
ANN_model.evaluate(x_test, y_test)
```

```python
# Lets go on with classification  report
from sklearn.metrics import confusion_matrix, classification_report
y_pred = ANN_model.predict(x_test)

y_pred_classes = [np.argmax(element) for element in y_pred]

print('Classification Report: \n', classification_report(y_test, y_pred_classes))
```

#### Using CNN for this dataset

```python
# diference in softmax and sigmoid
# In sigmoid you dont get sigmoid all the time
# you will get 2 in ans using sigmoid

# 1: 0.45
# 2: 0.65
    



# but in SOFTMAX it uses probability method
# like it will be
0.45/(0.45 + 0.65)
# 
# so softmax normalize the data with both value like
0.65/(0.65 + 0.45)

# both value sum is 1
```

```python
CNN_model = keras.Sequential([
    
#     first apply CNN 
# first set
    keras.layers.Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', input_shape = (32, 32, 3)),# filter give you for max accuracy with 32 features detection
    keras.layers.MaxPooling2D((2, 2)), 
    
# second set
    keras.layers.Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu'),# filter give you for max accuracy with 32 features detection
    keras.layers.MaxPooling2D((2, 2)), 
    
    
    
#     Dense
    keras.layers.Flatten(), #32 * 32 * 3 = 3072 so no of hiddel layers less than input
    keras.layers.Dense(64, activation = 'relu' ),

    keras.layers.Dense(10, activation = 'softmax' ) ,
    
# compiling model
])
```

```python
CNN_model.compile(optimizer = 'adam',  loss = 'sparse_categorical_crossentropy', metrics = ['accuracy']) 
CNN_model.fit(x_train, y_train, epochs = 10)
```
